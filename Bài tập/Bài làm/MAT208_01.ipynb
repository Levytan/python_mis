Cho thuật toán xác định số nghiệm phương trình bậc ba $ax^3 + bx^2 + cx + d = 0$ ($a \ne 0$) như sau:
1. Thay $t = x - \frac{b}{3a}$ vào phương trình $ax^3 + bx^2 + cx + d = 0$ 0 để được phương trình $t^3 +pt + q = 0$.2. Tính $\Delta = -(4p^3 + 27q^2)$.3. Tùy vào giá trị $\Delta$ mà kết luận số nghiệm
- $\Delta > 0$: phương trình có ba nghiệm
- $\Delta < 0$: phương trình có một nghiệm
- $\Delta = 0$ và $p = 0$: phương trình có một nghiệm.
- $\Delta = 0$ và $p \ne 0$: phương trình có hai nghiệm.

Viết chương trình python in ra số nghiệm của phương trình b ba ậc$ax^3 + bx^2 + cx + d = 0$ ($a \ne 0$) dựa vào thuật toán trên.

```python
def pt3(a, b, c, d):
    p = (3*a*c - b*b) / (3*a*a)
    q = (2*b**3 - 9*a*b*c + 27*a*a*d) / (27*a**3)
    delta = -(4*p**3 + 27*q**2)

    if delta > 0:
        return 1
    elif delta < 0:
        return 3
    else:
        if p == 0:
            return 1
        else:
            return 2
```

```python
pt3(1, 3, 3, 1)
```

Cho thuật toán xác định số dư phép chia đa thức $f(x) = \displaystyle\sum_{i = 0}^{n}a_i x^i$
cho $g(x) = x - \beta$ như sau:
\begin{equation*}
    \begin{cases}
        X_0 &= \beta \\
        X_{m+1} &= \beta X_m + a_{n-m-1}
    \end{cases}
\end{equation*}

Khi đó, $X_n$ chính là số dư cần tìm. Viết chương trình python cho thuật toán trên,
với hai tham số lần lượt là một danh sách ($[a_0,\ a_1,\ ...,\ a_n]$) và một số ($\beta$).

```python
def tìm_số_dự(A, beta):
    x = beta
    t = len(A)
    for m in range(1, len(A)):
        x = beta * x + A[t - m - 1]
    return x
```

```python
def in_số_may_mắn():
    for i in range(1000000):
        if set(str(i)) <= {'6', '8'}:
            print(i)
```

Cho một danh sách số A, cặp chỉ số (i, j) (i < j) được gọi là một nghịch thế nếu A[i] > A[j]. Viết chương trình đếm số nghịch thế trong một danh sách cho trước.

Ví dụ: Danh sách A = [7, 5, 4, 8] có 3 nghịch thế là (0, 1), (0, 2), (1, 2)

```python
def đếm_nghịch_thế(A):
    t = len(A)
    d = 0
    for i in range(t-1):
        for j in range(i+1, t):
            if A[i] > A[j]:
                d = d + 1

    return d
```

```python
A = [7, 5, 4, 8]
đếm_nghịch_thế(A)
```

Viết chương trình python tương ứng cho các định nghĩa đệ quy sau:
\begin{equation*}
    \textrm{gcd}(a, b) = \begin{cases}
        a &\text{nếu } a = b, \\
        \textrm{gcd}(b, a - b) &\text{nếu } a > b, \\
        \textrm{gcd}(a, b - a) &\text{nếu } a < b
    \end{cases}
\end{equation*}
và
\begin{equation*}
    {a\choose n} = \begin{cases}
        1 &\text{nếu } n = 0, \\
        {a-1\choose n} + {a-1\choose n-1}&\text{nếu } n > 0
    \end{cases}
\end{equation*}

```python
def gcd(a, b):
    if a == b:
        return a
    elif a > b:
        return gcd(b, a - b)
    else:
        return gcd(a, b - a)
```

```python
def choose(a, n):
    if n == 0:
        return 1
    else:
        return choose(a - 1, n) + choose(a - 1, n - 1)
```
