**Câu 2**: Với hai số tự nhiên $n$ và $b$ bất kỳ, ta định nghĩa hai dãy số $(u_i)$ và $(v_i)$ như sau:

$$(v_i): \left\{\begin{array}{l l l}
    v_0 &= &n \\
    v_{i+1} &= &(v_i - u_i)\ //\ b,\qquad i \geq 0
\end{array}
\right.
$$

$$
(u_i): u_{i} = v_i\ \%\ b, \qquad i \geq 0
$$
Viết chương trình nhập vào hai số $n$ và $b$, in ra màn các số $u_k$, $u_{k-1}$, $u_{k-2}$, $\ldots$, $u_0$ (các số cách nhau bằng dấu phẩy) với $k$ là số nhỏ nhất sao cho $v_{k+1} = 0$.

*Ví dụ*: Nhập $n = 9999$ và $b = 8$ sẽ in ra:
$$2, 3, 4, 1, 7$$

````python
n = int(input('Nhập n: '))
b = int(input('Nhập b: '))

d = []
v = n
while v > 0:
    u = v % b
    v = (v - u) // b
    d += [u]

for c in d[::-1]:
    print(c, end = ',')
````

**Câu 3**: Viết chương trình nhập vào một số tự nhiên $n$, in ra màn hình tổng các chữ số của số đó.

````python
n = int(input('Nhập n: '))

s = 0
m = n
while n > 0:
    s += n % 10
    n = n // 10
    
print('Tổng các chữ số của', m, 'là', s)
````

**Câu 5**: Số $m$ được định nghĩa là **'số giàu'** nếu tổng các ước của số đó **lớn hơn** hai lần số đó. Tương tự, số $n$ được định nghĩa là **'số nghèo'** nếu tổng các ước của số đó **nhỏ hơn** hai lần số đó. Viết chương trình tính số lượng số giàu và số nghèo nhỏ hơn 1000.

````python
def tổng_ước(n):
    s = 0

    for i in range(1, n):
        s += i if n % i == 0 else 0
    return s + n # vì n luôn là ước của n và vòng lặp kết thúc tại n-1

số_giàu = số_nghèo = 0
for i in range(1, 1000):
    c = tổng_ước(i)
    if c > 2 * i:
        số_giàu += 1
    elif c < 2 * i:
        số_nghèo += 1

print('Số lượng số giàu', số_giàu)
print('Số lượng số nghèo', số_nghèo)
````

**Câu 6**: Giả sử trong một lớp học có $n$ người, khi đó, xác suất để trong lớp học đó có ít nhất 2 người có cùng ngày sinh (không kể ngày 29.02) được tính bởi công thức:
$$
P = 1 - \frac{365}{365} \times \frac{364}{365} \times ... \times \frac{365-(n-1)}{365}
$$

Thực hiện 2 yêu cầu sau:
1. Viết chương trình nhập vào số $n$, in ra màn hình xác suất có ít nhất 2 người cùng ngày sinh.
2. Tìm số người tối thiểu để xác suất có 2 người cùng ngày sinh lớn hơn $50\%$.

````python
#6.1
n = int(input('Nhập n: '))

p = 1
for i in range(n):
    p *= (365-i)/365
    
print('Xác suất có ít nhất 2 người cùng ngày sinh là', '{:.2f}%'.format(100*(1-p)))
````

````python
#6.2
p = 1
i = 0
while p > 0.5:
    p *= (365-i)/365
    i += 1
    
print('số người tối thiểu để xác suất có 2 người cùng ngày sinh lớn hơn 50% là', i)
````

**Câu 7**: Viết chương trình nhập vào một chuỗi chỉ bao gồm các ký tự `a, t, c, và g` (phải kiểm tra chuỗi đã nhập). Đếm số lần xuất hiện của từng ký tự. Sau đó, với mỗi ký tự, in ra màn hình câu sau:

```Ký tự … xuất hiện … lần.```

Ví dụ: Nhập vào chuỗi `atctccacct` sẽ in ra:

```
Ký tự a xuất hiện 2 lần.
Ký tự t xuất hiện 3 lần.
Ký tự c xuất hiện 5 lần.
Ký tự g xuất hiện 0 lần.
```

````python
s = input('Nhập chuỗi ')
adn = ['a', 'c', 't', 'g']
while set(s) != set(adn): # set(s) sẽ trả về những ký tự xuất hiện trong chuỗi s
    s = input('Nhập chuỗi ')
    
d = {
    'a': 0, 't': 0, 'c': 0, 'g': 0
}

for c in s:
    d[c] += 1
    
for c in adn:
    print(f'Ký tự {c} xuất hiện {d[c]} lần.')
````

**Câu 8**: Viết chương trình nhập vào một ngày tháng dạng $dd/mm$ (ngày trước tháng sau), in ra màn hình cách ghi ngày đó trong tiếng Anh theo dạng `tên tháng, thứ tự ngày`.

Ví dụ: nhập vào $07/03$ sẽ in ra `March, 7th`.

````python
s = input('Nhập chuỗi dạng dd/mm ')

d, m = s.split('/')

month = {
    1: 'January', 2: 'February', 3: 'March', 4: 'April',
    5: 'May', 6: 'June', 7: 'July', 8: 'August',
    9: 'Septemper', 10: 'October', 11: 'November', 12: 'December'
}
d = int(d)
if d in (1, 21, 31):
    d = str(d) + 'st'
elif d in (2, 22):
    d = str(d) + 'nd'
elif d in (3, 23):
    d = str(d) + 'rd'
else:
    d = str(d) + 'th'
    
print(month[int(m)] + ', ' + d)
````

**Câu 9**: Hai số $(m, n)$ được gọi là hai số bạn bè nếu tổng các ước (không tính chính số đó) của một số bằng với số còn lại.

*Ví dụ*: $220$ và $284$ là 2 số bạn bè, vì ta có tổng các ước của $220\ (1, 2, 4, 5, 10, 11, 20, 22, 44, 55, 110)$ là 284 và tổng các ước của $284\ (1, 2, 4, 71, 142)$ là 220.

Viết chương trình in ra màn hình các cặp số bạn bè $(a, b)$ với $a \leq 10000$.

````python
def tổng_ước(n):
    s = 0

    for i in range(1, n):
        s += i if n % i == 0 else 0
    return s

for a in range(1, 10001):
    b = tổng_ước(a)
    if a < b and tổng_ước(b) == a:
        print(a, b)
````

**Câu 10**: Viết chương trình nhập vào một chuỗi. Đếm số lần xuất hiện của từng ký tự *(không phân biệt hoa thường, chỉ tính các ký tự là chữ cái hoặc chữ số)*. Sau đó, với mỗi ký tự xuất hiện trong chuỗi đã nhập, in ra màn hình câu sau:

```Ký tự … xuất hiện … lần.```

Ví dụ: Nhập vào chuỗi Hello world! sẽ in ra:

```
Ký tự h xuất hiện 1 lần.
Ký tự e xuất hiện 1 lần.
Ký tự l xuất hiện 3 lần.
Ký tự o xuất hiện 2 lần.
Ký tự w xuất hiện 1 lần.
Ký tự r xuất hiện 1 lần.
Ký tự d xuất hiện 1 lần.
```

````python
s = input('Nhập chuỗi ')

s = s.lower() #s.lower() sẽ trả về chuỗi được viết thường
d = {c for c in s if c.isalnum()} # c.isalnum() trả về True nếu c là ký tự chữ cái hoặc ký tự số

count = {c: 0 for c in d} # khởi tạo biến đếm

for c in s:
    if c in d:
        count[c] += 1
    
for c in d:
    print(f'Ký tự {c} xuất hiện {count[c]} lần.')
````

**Câu 11**: Viết chương trình nhập vào một số nguyên $n$ $(0 ≤ n ≤ 99)$, in ra cách đọc của số đó trong tiếng Việt. Nếu không thỏa điều kiện, yêu cầu nhập lại.

````python
n = int(input('Nhập n: '))

while not(0 <= n <= 99):
    n = int(input('Nhập n: '))
    
đếm = {
    0: 'không', 1: 'một', 2: 'hai', 3: 'ba', 4: 'bốn',
    5: 'năm', 6: 'sáu', 7: 'bảy', 8: 'tám', 9: 'chín'
}

đơn_vị = n % 10
chục = n // 10

if chục == 0:
    đọc = đếm[đơn_vị]
elif chục == 1:
    đếm[0] = ''
    đếm[5] = 'lăm' # đổi cách đọc cho hai số 0 và 5
    đọc = 'mười ' + đếm[đơn_vị]
else:
    đọc = đếm[chục] + ' mươi '
    đếm[0] = ''
    đếm[5] = 'lăm' # đổi cách đọc cho hai số 0 và 5
    đọc = đọc + đếm[đơn_vị]

print(đọc)
````
